```python
import pickle
import numpy as np
import itertools
import mne
results = pickle.load( open( "C:\\Users\\nemrodov\\Documents\\Ilya_study\\Analysis\\svm_perc_coefs.pkl", "rb" ) )
epochs = mne.read_epochs( "C:\\Users\\nemrodov\\Documents\\Ilya_study\\Analysis\\S15_perc_aver-epo.fif")
results = np.array(results)
combs = list(itertools.combinations(np.arange(100), 2))
```

```python
def find_index_combs(numset,combs):
    import numpy as np
    numset = set(numset)
    indx = []
    for i,j in enumerate(combs):
        if len(numset.intersection(set(j))) == 2:
            indx.append(i)
    return indx

```

```python
import matplotlib.pyplot as plt
import numpy as np
import mne

fig, ax = plt.subplots(figsize=(7.5, 10),nrows=2,ncols=2)

numset = np.arange(0,25)
indx = find_index_combs(numset,combs)
a = np.mean(results[:,:,indx,:],axis = (1,2)).squeeze()
a = np.abs(a)
mne.viz.plot_topomap(a,epochs.info,vmin=np.min(a),axes=ax[0,0], show=False)
ax[0,0].set_title('upright unfamiliar', fontsize=14)


numset = np.arange(25,50)
indx = find_index_combs(numset,combs)
a = np.mean(results[:,:,indx,:],axis = (1,2)).squeeze()
a = np.abs(a)
mne.viz.plot_topomap(a,epochs.info,vmin=np.min(a),axes=ax[0,1], show=False)
ax[0,1].set_title('upright famous', fontsize=14)

numset = np.arange(50,75)
indx = find_index_combs(numset,combs)
a = np.mean(results[:,:,indx,:],axis = (1,2)).squeeze()
a = np.abs(a)
mne.viz.plot_topomap(a,epochs.info,vmin=np.min(a),axes=ax[1,0], show=False)
ax[1,0].set_title('inverted unfamiliar', fontsize=14)

numset = np.arange(75,100)
indx = find_index_combs(numset,combs)
a = np.mean(results[:,:,indx,:],axis = (1,2)).squeeze()
a = np.abs(a)
mne.viz.plot_topomap(a,epochs.info,vmin=np.min(a),axes=ax[1,1], show=False)
ax[1,1].set_title('inveted famous', fontsize=14)

plt.show()
```
